### Correlation Matrix

```python
corr=advertising.corr()
corr
```

### Plot the Correlation Matrix

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(9, 7))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### Figure 3.5 Multiple variable regression

```python
regr=skl_lm.LinearRegression()
```

```python
X=advertising[['radio','TV']].as_matrix() #select the features
```

```python
Y=advertising.sales
```

```python
regr.fit(X,Y)# Run the regression
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```

```python
# What are the min/max values of Radio & TV?
# Use these values to set up the grid for plotting.
advertising[['radio','TV']].describe()
```

```python
#Create a coordinate grid
Radio=np.arange(0,50)
TV=np.arange(0,300)
B1, B2=np.meshgrid(Radio, TV, indexing='xy') #Return coordinate matrices from coordinate vectors.
Z=np.zeros((TV.size, Radio.size)) #返回来一个给定形状和类型的用0填充的数组
for (i,j), v in np.ndenumerate(Z):#Return an iterator yielding pairs of array coordinates and values.
    Z[i,j]=(regr.intercept_+B1[i,j]*regr.coef_[0]+B2[i,j]*regr.coef_[1])
```

```python
#Create plot
fig=plt.figure(figsize=(10,6))##生成一个空图
fig.suptitle('Regression: Sales ~ Radio + TV Advertising', fontsize=20, color='blue')#设置图片标题

ax=axes3d.Axes3D(fig)
#Create a surface plot.
ax.plot_surface(B1,B2, Z, rstride=10, cstride=5, alpha=0.5)
ax.scatter3D(advertising.radio, advertising.TV, advertising.sales, c='r')
```
